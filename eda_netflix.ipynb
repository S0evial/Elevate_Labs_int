```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\vishw\OneDrive\Desktop\INTSHP\task 1\netflix_titles_raw.csv', encoding = 'latin1')
df.head()
```

```python
# Finding and handling Missing Values
missing_value = df.isnull().sum()
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not listed')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].fillna(pd.Timestamp('2000-01-01'))
#print(missing_value)
print(df.shape) 
print(df.isnull().sum())
```

```python
#Dropping Unnamed Column
df = df.drop(columns=[col for col in df.columns if 'Unnamed' in col])
```

```python
#Removing Duplicate Values
df.drop_duplicates(inplace = True)
```

```python
#Standardixing Texts
df['country'] = df['country'].str.lower().str.strip()
df['director'] = df['director'].str.title().str.strip()

#Renameing Columns name 
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# Date and Time formatting
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
print(df.shape)
```

```python
#saving the data set
df.to_csv('Netflix_cleaned.csv', index = False)
```
